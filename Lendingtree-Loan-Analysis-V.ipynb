### 📊 Exploratory Data Analysis & Visualisations using `Matplotlib Subplots`

Just like we can stack commands from both Seaborn and Matplotlib to create a visually appealing chart, we can also stack multiple plots on top of each other. 

We may want to use this technique to:
* Stack complementary plots or
* Plot multiple distributions together on the same axes. 

## Scenario - 1:
### `Rug Plot` with Histograms

* One limitation of **histograms** is that they obscure the mass of data within the bins, so **we lose sight of the individual values.**

* So, a **histogram** is an **aggregated plot**, and it's **complemented well by** a **disaggregated plot** - one that **shows all the individual values.** 

* One option we have is a `rug plot`.

* A **rugplot** displays all the individual values on the specified axis, typically at the base of the histogram, so we get the benefit of visualizing the shape of the distribution and the location of the individual data points.

* This is a Seaborn plot function. We can just stack this plot right on top of the previous plot.

**Goal:**

* We are interested in characterizing the different features present in the dataset. It contains both:
    
    * Features of **each person**, like their **occupation** and **loan history**, and
    
    * Features of the **individual loans** in the dataset, like the **loan amount** and **interest rate.**

**Task:**

Based on some previous analysis, we know that the **loan amount** feature in our dataset has some interesting patterns. 
We want to visualize this data using a **histogram** and **rugplot** to explain its unique properties to our clients. 
This visualization will help them identify and price common loan products based on what their competitors may offer.

## Scenario - 3:
### Overlaying `Histograms` on Same Chart

We may be interested in plotting multiple histograms together on the same axes to compare distributions. 

* For example, if we're interested in comparing the distribution of loan amounts for A and D grade loans, we can first filter these into two separate data frames: A is only for when the grade matches A, and D is only for when the grade matches D.
  
* Then, we can just go ahead and reuse the histogram code, changing this data frame to A's only, plus add another hist plot. Just like we stacked the rug plot on top here, we can just stack another histogram here too, this time for D's only. We can keep all the same parameters except the color, which you can keep as the default blue for the second histogram.

*So we can see the D loans have a less skewed distribution, and for the most part, A loans appear to be more common since there are only a few bins where D loans outnumber A's in the 25,000 to 30,000 range.*

* Since we have two plots, we could call `plot.show()` twice to see each of them individually if we like.
